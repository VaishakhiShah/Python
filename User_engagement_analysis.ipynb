user_id: Unique identifier for the user
region: The user's region (e.g., North America, Europe, Asia, etc.)
post_views: Number of times the user's posts were viewed
likes_received: Number of likes received on the user's posts
comments_received: Number of comments received on the user's posts
engagement_score: A calculated score based on views, likes, and comments
Import the data set from this URL:

http://becomingvisual.com/python4data/user_engagement.csv

## Task Overview:

Perform specific data manipulations to focus on the European market, and then visualize the engagement patterns for this region.



## Analysis

# Setup and Data Loading:


```python
# Import the data

import pandas as pd
engagement_df = pd.read_csv("http://becomingvisual.com/python4data/user_engagement.csv", header=0)

# Showing first 15 rows
print(engagement_df.head(15))
print(engagement_df.shape)
```

# Data Filtering and Processing:

Filtering engagement_df to retain only the data for users located in "Europe". Store this filtered data in a DataFrame called europe_engagement_df.
For the "Europe" data, create a new column interaction_rate calculated as the sum of likes_received and comments_received divided by post_views, multiplied by 100. This metric will gauge the interaction rate per view.

```python
# Data Filtering from dataframe using .isin

europe_engagement_df = engagement_df[engagement_df['region'].isin(['Europe'])].copy()
europe_engagement_df['interaction_rate'] = ((europe_engagement_df['likes_received'] + europe_engagement_df['comments_received']) / europe_engagement_df['post_views']) * 100

print(europe_engagement_df)
```

# Descriptive Statistics:

Calculating and displaying the mean, median, and standard deviation of the engagement_score for users in Europe.
Identify the user_id of the user with the maximum interaction_rate in Europe and display it.

```python
#preview with describe()
europe_engagement_df.describe()
```

```python
## Calculating the descritive stats

# mean,median, and standard deviation of the engagement_score for users in Europe (rounding to 2 decimal places)

mean_eng_score =round(europe_engagement_df["engagement_score"].mean(),2)
median_eng_score =round(europe_engagement_df["engagement_score"].median(),2)
std_dev_eng_score =round(europe_engagement_df["engagement_score"].std(),2)

# Displaying the descriptive stats

print("The mean of the engagement score for users in Europe:",mean_eng_score)
print("The median of the engagement score for users in Europe:",median_eng_score)
print("The median of the engagement score for users in Europe:", std_dev_eng_score)

# Identify the user_id of the user with the maximum interaction_rate in Europe and display it using idmax

user_id_max_rate = europe_engagement_df.loc[europe_engagement_df["interaction_rate"].idxmax()]
print("The user id of the user with the maximum interaction rate in Europe:", user_id_max_rate["user_id"])
```

# Data Visualization:

Using Pandas Plot or Matplotlib , creating a histogram to visualize the distribution of engagement_score for the European users. Ensure the histogram has proper labels and a title.
Generating a scatter plot to illustrate the relationship between likes_received and interaction_rate for the European region. Include axis labels, a title, and any relevant annotations to highlight significant insights.

```python
# Importing Matplotlib

import matplotlib.pyplot as plt
```

```python
# Histogram to visualize the distribution of engagement_score for the European users

#create a Figure
plt.figure(figsize=(8, 6))
plt.hist(europe_engagement_df["engagement_score"], bins=15, color='#572C86', edgecolor='black')

#adding title and axis lables
plt.title(" Engagement Scores Distribution for European Users")
plt.xlabel("Engagement Score")
plt.ylabel("Frequency")

#remove ticks and grids
plt.tick_params(axis='both', which='both', length=0)
plt.grid(False)

#showing the plot

plt.show()
```

```python
# Scatter plot to illustrate the relationship between likes_received and interaction_rate for the European region.

#create a Figure
plt.figure(figsize=(8, 6))
plt.scatter(europe_engagement_df["likes_received"], europe_engagement_df["interaction_rate"], color='#4cbea3', alpha=0.7)

#adding title and axis lables

plt.title("Likes Received versus Interaction Rate for European Users")
plt.xlabel("Likes Received")
plt.ylabel("Interaction Rate")

#remove ticks and grids
plt.tick_params(axis='both', which='both', length=0)
plt.grid(False)
plt.tight_layout()

#showing the plot
plt.show()
```
