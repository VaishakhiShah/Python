#1. Import data

```
import pandas as pd
taxi = pd.read_csv("http://becomingvisual.com/python4data/taxi2016.csv", header=0)
print(taxi.head(5))
print(taxi.shape)
```

#2.Subset data to only include those trips with 1 passenger

```
#taxi_subset = taxi[taxi.passenger_count == 1]
taxi_subset = taxi.loc[taxi.passenger_count == 1].copy()
taxi_subset.head(5)
```

```
taxi_subset.shape[0]
```

```
taxi_subset.info()
```

```
#Reset index
taxi_subset.reset_index(drop=True, inplace=True)
```

```
taxi_subset.info()
```

# 3. Calculate the average `tip_amount`

```
#Option 1.
taxi_tip_avg_pass01 = sum(taxi_subset.tip_amount) / taxi_subset.shape[0]
taxi_tip_avg_pass01
```

```
#Option 2. Use the .mean() method
taxi_subset.tip_amount.mean()
```

# 4. Calculate the average `fare_amount`

```
# Calculate the average fare_amount
taxi_subset.fare_amount.mean()
```

#5. Find the range, min, and max of tips

```
#preview with describe()
taxi_subset.describe()
```

```
print("Maximum tip was:",taxi_subset.tip_amount.max() )
print("Minimum tip was:", taxi_subset.tip_amount.min())
print("The range is:", taxi_subset.tip_amount.min(), ":",taxi_subset.tip_amount.max())
```

#6. Group by passenger count and use .size() to show how many passengers are in each group

```
taxi.groupby(['passenger_count']).size()
```

#7.  Compute the mean tip amount, total fare, and distance traveled.

```
taxi.groupby(by=['payment_type'])[['total_amount', 'tip_amount',  'trip_distance']].mean()
```
