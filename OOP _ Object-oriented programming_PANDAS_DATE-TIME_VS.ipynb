```
# we can add the new attribute in the BookBuilder: Version 3.0 object

## Lets say we added new def below then re-run everything


class BookBuilder:

  def __init__(self, author, title, publisher):
      print ("Hey, you've created an object of type BookBuilder!")
      self.title = title
      self.author = author
      self.publisher = publisher
      print("The title is:", self.title)
      print("The author is:", self.author)
      print("The publisher is:", self.publisher)

  def setPrice (self, price):
      self.price=price
      print("The price is", self.price)

  def setPages (self,pages):
      self.pages = pages
      print("The pages are", self.pages)
```
