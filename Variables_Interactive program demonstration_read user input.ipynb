### X. The challenge:

Create a program that asks a user for their netid and number of hours they spent on lesson 1 in this course. Then print out a statement that multiples that number of remaining lessons (11) by the number hours spent on lesson 1.

#### Approach 1

```
id =input("What is your netid? \n") # adding new line
hours = input("how many hours did you spend on lesson 1?\n") # adding new line
print("There are 11 more lessons. You should expect to spend at least", 11
      * float(hours), "hours between now and the last class.")
```

#### Approach 2

```
#adding exception

id =input("What is your netid? \n") # adding new line
hours = input("how many hours did you spend on lesson 1?\n") # adding new line

try:
  hours=float(hours)
  print("There are 11 more lessons. You should expect to spend at least", 11 * hours, "hours between now and the last class.")
except:
  print("Please start over and enter a number.")
```

#### Approach 3

```

#using while loop

invalid = True
id =input("What is your netid? \n") # adding new line

while invalid:

  hours = input("how many hours did you spend on lesson 1?\n") # adding new line

  try:

    hours=float(hours)
    print("There are 11 more lessons. You should expect to spend at least", 11 * hours, "hours between now and the last class.")
    invalid = False
    #also could use a break here
  except:
    print("Your input:", hours, "is invalid. Please start over and enter a number.")
```

## Write your name in binary.

```
myname ="Kristen"
ord(myname[0])
format(ord(myname[0]), '08b') #displays just K
```

```
# initializing string
test_str = "Kristen"

# printing original string
print("The original string is : " + str(test_str))

# using join() + ord() + format()
# Converting String to binary
res = ' '.join(format(ord(i), '08b') for i in test_str) #advanced topic on for loops. See lesson 5

# printing result
print("The string after binary conversion : " + str(res))
```
