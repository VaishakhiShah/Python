# Nested Dictionary

## dictionary within dictionary

```python
mygarage = { "spot01":{"brand": "Tesla","model":"Model X","year": 2020}, # the Keys have dictionary as a value
            "spot02": {"brand": "Tesla","model": "Model Y","year": 2019}}

print(mygarage)
```

```python
mygarage
```

```python
mygarage["spot01"]
```

```python
mygarage["spot01"]["year"]
```

```python
mygarage["spot02"]["brand"] #dual indexing
```

```python
mygarage.values()
```

```python
mygarage.keys()
```

```python
mygarage["spot01"].keys()
```

```python
mygarage["spot02"].keys()
```

```python
mygarage["spot01"].values()
```

```python
mygarage["spot02"].values()
```

```python
mygarage.items()
```

```python
mygarage["spot01"].items()
```

```python
for k, v in mygarage.items(): # For loop : use it when we have finite no. of items -- k,v here b/w for....in
  print(k)
  print(v)
print("Done")
```

```python
for k in mygarage["spot01"].keys():
  print(k)
print("Done")
```

```python
for v in mygarage.values():
  print(v)
print("Done")
```

```python
charity = { "donate01":{"product": "oldphone","type":"iphone","year": 2016}, # the Keys have dictionary as a value
            "donate02": {"product": "book","type": "comic","year": 2012}}
```
